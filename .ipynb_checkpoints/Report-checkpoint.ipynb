```python
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
entries = os.listdir('./lin318/')
results = pd.DataFrame()

for entry in entries:
    new_data = pd.read_csv('./lin318/' + entry, delimiter=',',header = None, error_bad_lines= False)
    results = pd.concat([results, new_data])
    
```

```python

#cols = gen, max, avg, min

means = results.groupby(results.index).mean()
#max
#print(means[1])
#avg
#print(means[2])
labels= ['avg of max', 'avg of avg']
fig, ax = plt.subplots()
ax.plot(means.index, means[1])
ax.plot(means.index, means[2])
ax.legend(labels)
ax.set_title('lin318, p_xover: 0.7, p_mut:0.5, popSize = 4')

```

```python
results[results.index == 10000].count()[0]
```

```python
#stats
n_trials = results[results.index == 10000].count()[0]
max_fitness = 9957971

avg_best_fitness = (results[results.index == 10000][7].mean())
opt_path_length = 10000 - max_fitness
avg_path_length = 10000 - avg_best_fitness

quality = 1-(avg_path_length - opt_path_length)/opt_path_length
print ("quality:", quality)

reliable_results = (results[results.index == 10000][7]== avg_best_fitness).count()

reliability = reliable_results/n_trials

print("reliability: ", reliability)

generation_speed =  means[means[1]==avg_best_fitness].idxmin()[7]
eval_speed = 2* generation_speed
print("speed: ", eval_speed)
```
